### Make a function that analyzes each test columnm and creates a new bool column detailing whether the failed test was the scale registered or not

```python
def isAllWeightTest(row):
    cols = ['Test 0 lbs', 'Test 50 lbs', 'Test 100 lbs', 'Test 150 lbs', 'Test 200 lbs', 'Test 250 lbs', 'Test 300 lbs', 'Test 350 lbs', 'Test 400 lbs', 'Test 450 lbs', 'Test 500 lbs', 'Test 550 lbs', 'Test 600 lbs', 'Test Second 0 lbs']
    for col in cols:
        if row.get(col) == 'Fail':
            return False
    return True

df['isAllWeightTestPass'] = df.apply(isAllWeightTest, axis=1)
df
```

### How many scales failed at least one weight test?

```python
df['isAllWeightTestPass'].value_counts()

# True: 249 scales passed every weight test it took
# False: 135 scales failed at least one weight test
```
